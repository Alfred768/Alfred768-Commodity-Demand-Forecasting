```python
# 多些特征
def create_date_features(df):
    df['month'] = df.date.dt.month
    df['day_of_month'] = df.date.dt.day
    df['day_of_year'] = df.date.dt.dayofyear
    df['week_of_year'] = df.date.dt.weekofyear
    df['day_of_week'] = df.date.dt.dayofweek + 1
    df['year'] = df.date.dt.year
    df["is_wknd"] = df.date.dt.weekday // 4
    df['is_month_start'] = df.date.dt.is_month_start.astype(int)
    df['is_month_end'] = df.date.dt.is_month_end.astype(int)
    return df
df = create_date_features(df)
df.head()
```

```python
def random_noise(dataframe):
    return np.random.normal(size=(len(dataframe),))
```

```python
# the feature name will be created dynamically with regards to the lag value for a given list of lags
def lag_features(dataframe, lags):
    dataframe = dataframe.copy()
    for lag in lags:
        dataframe['sales_lag_' + str(lag)] = dataframe['sales'].transform(
            lambda x: x.shift(lag)) + random_noise(dataframe)
    return dataframe

df = lag_features(df, [91, 98, 105, 112, 119, 126, 182, 364, 546, 728])
df.head()
```

```python
def roll_mean_features(dataframe, windows):
    dataframe = dataframe.copy()
    for window in windows:
        dataframe['sales_roll_mean_' + str(window)] = dataframe['sales'].transform(
            lambda x: x.shift(1).rolling(window=window, min_periods=10, win_type="triang").mean()
        ) + random_noise(dataframe)
    return dataframe
df = roll_mean_features(df, [365, 546, 730])
df.head()
```

```python
def ewm_features(dataframe, alphas, lags):
    dataframe = dataframe.copy()
    for alpha in alphas:
        for lag in lags:
            dataframe['sales_ewm_alpha_' + str(alpha).replace(".", "") + "_lag_" + str(lag)] = \
                dataframe['sales'].transform(lambda x: x.shift(lag).ewm(alpha=alpha).mean())
    return dataframe

alphas = [0.95, 0.9, 0.8, 0.7, 0.5]
lags = [91, 98, 105, 112, 180, 270, 365, 546, 728]
df = ewm_features(df, alphas, lags)
df.head()
```

```python
df = pd.get_dummies(df, columns=['day_of_week', 'month'])
df.head()
```

```python
df['sales'] = np.log1p(df["sales"].values)
df['sales'].head()
```

```python
def smape(preds, target):
    n = len(preds)
    masked_arr = ~((preds == 0) & (target == 0))
    preds, target = preds[masked_arr], target[masked_arr]
    num = np.abs(preds-target)
    denom = np.abs(preds)+np.abs(target)
    smape_val = (200*np.sum(num/denom))/n
    return smape_val

def lgbm_smape(preds, train_data):
    labels = train_data.get_label()
    smape_val = smape(np.expm1(preds), np.expm1(labels))
    return 'SMAPE', smape_val, False
```

```python
train = df.loc[(df["date"] < "2017-07-01"), :]
train["date"].min(), train["date"].max()
```

```python
val = df.loc[(df["date"] >= "2017-07-01") & (df["date"] < "2017-10-01"), :]
val["date"].min(), val["date"].max()
```

```python
cols = [col for col in train.columns if col not in ['date',"sales", "year"]]
```

```python
Y_train = train['sales']
X_train = train[cols]

Y_val = val['sales']
X_val = val[cols]
```

```python
lgb_params = {'metric': {'mae'},
              'num_leaves': 32,
              'learning_rate': 0.02,
              'feature_fraction': 0.8,
              'max_depth': 16,
              'verbose': 0,
              'num_boost_round': 1000,
              'early_stopping_rounds': 300,
              'subsample': 0.8,
              'nthread': -1}
```

```python
lgbtrain = lgb.Dataset(data=X_train, label=Y_train, feature_name=cols)
lgbval = lgb.Dataset(data=X_val, label=Y_val, reference=lgbtrain, feature_name=cols)
```

```python
model = lgb.train(lgb_params, lgbtrain,
                  valid_sets=[lgbtrain, lgbval],
                  num_boost_round=lgb_params['num_boost_round'],
                  early_stopping_rounds=lgb_params['early_stopping_rounds'],
                  feval=lgbm_smape,
                  verbose_eval=200)
```

```python
y_pred_val = model.predict(X_val)
```

```python
smape(np.expm1(y_pred_val), np.expm1(Y_val))
```

```python
plt.figure(figsize=(12,6))
plt.plot(np.expm1(y_pred_val),color='blue',label='pred')
plt.plot(np.expm1(Y_val.values),color='red',label='true')
plt.legend()
plt.show()
```

```python
train = df.loc[~df.sales.isna()]
Y_train = train['sales']
X_train = train[cols]

test = df.loc[df.sales.isna()]
X_test = test[cols]
```

```python
lgb_params = {'metric': {'mae'},
              'num_leaves': 32,
              'learning_rate': 0.02,
              'feature_fraction': 0.8,
              'max_depth': 16,
              'verbose': 0,
              'nthread': -1,
              'subsample': 0.8,
              "num_boost_round": model.best_iteration}
```

```python
lgbtrain_all = lgb.Dataset(data=X_train, label=Y_train, feature_name=cols)
final_model = lgb.train(lgb_params, lgbtrain_all, num_boost_round=model.best_iteration)
test_preds = final_model.predict(X_test, num_iteration=model.best_iteration)
```

```python
test_df['sales'] = np.expm1(test_preds)
test_df.to_csv('lgb.csv',index_label=False,index=False)
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# 读入数据集
df = pd.read_csv('train.csv')
df.drop(columns=['store','item'],inplace=True)
df.columns = ['ds','y']
print(df.head())
# 拟合模型
m = Prophet()
m.fit(df)

# 构建待预测日期数据框，periods = 365 代表除历史数据的日期外再往后推 365 天
future = m.make_future_dataframe(periods=92)
future.tail()
# 预测数据集
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
m.plot(forecast)
m.plot_components(forecast)
plt.show()
```

```python
prophet_pred = forecast.iloc[-92:,:][['ds','yhat']].reset_index(drop=True).rename(columns={'ds':'date','yhat':'sales'})
prophet_pred.to_csv('prophet.csv',index_label=False,index=False)
```
